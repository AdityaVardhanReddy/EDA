```python
Data_0107_cleaned = Data_0107.dropna()
print(f"Number of rows after dropping missing values: {len(Data_0107_cleaned)}")
```

```python
for column in ['job', 'education', 'contact', 'poutcome']:
    Data_0107[column].fillna(Data_0107[column].mode()[0], inplace=True)

print(Data_0107.isnull().sum())
Data_0107.head()
```

```python
duplicates = Data_0107.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

Data_0107_cleaned = Data_0107.drop_duplicates()

print(f"Number of rows after removing duplicates: {len(Data_0107_cleaned)}")
```

```python
# Define age bins and labels
age_bins = [0, 18, 65, 100]
age_labels = ['Child', 'Adult', 'Senior']

# Create a new column 'age_group' based on the bins
Data_0107['age_group'] = pd.cut(Data_0107['age'], bins=age_bins, labels=age_labels)

# Display the updated DataFrame
Data_0107[['age', 'age_group']].head()
```

```python
# Define balance bins and labels
balance_bins = [-float('inf'), 0, 1000, float('inf')]  # Categories: Negative/Zero, Low, High
balance_labels = ['Negative/Zero', 'Low', 'High']

# Create a new column 'balance_category' based on the bins
Data_0107['balance_category'] = pd.cut(Data_0107['balance'], bins=balance_bins, labels=balance_labels)

# Display the updated DataFrame
Data_0107[['balance', 'balance_category']].head()
```

```python
# Define duration bins and labels
duration_bins = [0, 200, 600, float('inf')]  # Short: 0-200, Medium: 200-600, Long: 600+
duration_labels = ['Short', 'Medium', 'Long']

# Create a new column 'duration_category' based on the bins
Data_0107['duration_category'] = pd.cut(Data_0107['duration'], bins=duration_bins, labels=duration_labels)

# Display the updated DataFrame
Data_0107[['duration', 'duration_category']].head()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Set plot aesthetics
sns.set(rc={'figure.figsize': (11, 6)})
plt.rcParams['figure.dpi'] = 150
# Select only numeric columns for analysis
numeric_data = Data_0107.select_dtypes(include=[np.number])
```

```python
# Mean
mean_values = numeric_data.mean()
print("Mean Values:")
print(mean_values)
# Median
median_values = numeric_data.median()
print("\nMedian Values:")
print(median_values)
# Quantiles
quantiles = numeric_data.quantile([0.25, 0.5, 0.75])
print("\nQuantiles:")
print(quantiles)
```

```python
# Range
range_values = numeric_data.max() - numeric_data.min()
print("\nRange:")
print(range_values)
# IQR
IQR = numeric_data.quantile(0.75) - numeric_data.quantile(0.25)
print("\nIQR:")
print(IQR)
# Interdecile Range (90th - 10th percentile)
interdecile_range = numeric_data.quantile(0.90) - numeric_data.quantile(0.10)
print("\nInterdecile Range:")
print(interdecile_range)
# Mean and Standard Deviation
mean = numeric_data.mean()
std_dev = numeric_data.std()
print("\nMean and Standard Deviation:")
print("Mean:\n", mean)
print("Standard Deviation:\n", std_dev)
# Skewness and Kurtosis
skewness = numeric_data.skew()
kurtosis = numeric_data.kurt()
print("\nSkewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
```

```python
# Frequency Distribution for numeric variables
numeric_data = Data_0107.select_dtypes(include=['int64', 'float64'])
frequency_distribution = numeric_data.apply(pd.Series.value_counts).fillna(0)
print("Frequency Distribution:")
print(frequency_distribution)
```

```python
# Histogram for numeric variables
numeric_data.hist(figsize=(15, 10), bins=10)
plt.suptitle("Histograms of Numeric Variables\n21BDS0107 Adityavardhan Reddy Papudippu")
plt.show()
```

```python
# Relative Frequency Distribution for numeric variables
relative_frequency = numeric_data.apply(lambda x: x.value_counts(normalize=True)).fillna(0)
print("Relative Frequency Distribution:")
print(relative_frequency)
```

```python
# Cumulative Frequency Distribution for numeric variables
cumulative_frequency = numeric_data.apply(lambda x: x.value_counts().cumsum()).fillna(0)
print("Cumulative Frequency Distribution:")
print(cumulative_frequency)

# Cumulative Frequency Plot
cumulative_frequency.plot(figsize=(15, 10), marker='o')
plt.title("Cumulative Frequency Distribution\n21BDS0107 Adityavardhan Reddy Papudippu")
plt.xlabel("Values")
plt.ylabel("Cumulative Frequency")
plt.grid()
plt.show()
```
